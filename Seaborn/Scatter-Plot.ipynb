```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Salary_Data.csv")
df.head()
```

```python
sns.scatterplot(x='YearsExperience',y='Salary',data=df);
```

```python
sns.scatterplot(x='YearsExperience',y='Salary',data=df, hue="Gender");
```

```python
sns.scatterplot(x='YearsExperience',y='Salary',data=df, hue="Gender", size="YearsExperience");
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='YearsExperience',y='Salary',data=df, hue="Gender", size="YearsExperience");
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='YearsExperience',y='Salary',data=df, hue="Gender", size="YearsExperience").set(title="Correlation between Salary and Years of Experience");
```
