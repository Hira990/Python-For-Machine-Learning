```python
import pandas as pd
import numpy as np 
df = pd.read_csv('housing.csv')
```

```python
df.head()
```

```python
def change_unit(num):
  return num/100000

df['median_house_value'] = df['median_house_value'].apply(change_unit)
```

```python
df.head()
```

```python
def get_income_group(income):
  if income >= 5:
    return "High"
  elif income < 5 and income > 2.5:
    return "Middle"
  else:
    return "Low"

df['income_group'] = df['median_income'].apply(get_income_group)
df
```

```python
def get_ratio(bedrooms, total_rooms):
  return bedrooms / total_rooms 

df['ratio'] = df[['total_rooms', 'total_bedrooms']].apply(lambda df: get_ratio(df['total_bedrooms'], df['total_rooms']), axis=1)
df
```

```python
df['ratio'] = np.vectorize(get_ratio)(df['total_bedrooms'], df['total_rooms'])
df
```
